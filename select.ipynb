```python
#We can observed that nost of the Customers are from UK. So, we further drill down to UK
uk_df=df[df.Country=='United Kingdom']
uk_df
```

```python
uk_df.describe()
```

```python
#from the above we can see that some customers have ordered in a negative value which is not possible. so we need to filter Quantity >0 

uk_df = uk_df[(uk_df['Quantity']>0)]
uk_df.describe()
```

```python
#The invoice Dates helps us to calculate Recency, Invoice number helps to count the number of transaction (Frequency) & Quantity
#purchase of each transaction *unit price calculate the Monetary value

uk_df = uk_df[['CustomerID', 'InvoiceDate', 'InvoiceNo', 'Quantity', 'UnitPrice']]
uk_df['TotalPrice'] = uk_df['Quantity'] *uk_df['UnitPrice']
uk_df['InvoiceDate'].min(),uk_df['InvoiceDate'].max()
PRESENT = dt.datetime(2011,12,10)
uk_df['InvoiceDate'] = pd.to_datetime(uk_df['InvoiceDate'])

```

```python
uk_df
```

```python
### RFM Analysis ###

rfm = uk_df.groupby('CustomerID').agg({'InvoiceDate': lambda date: (PRESENT - date.max()).days,
'InvoiceNo': lambda num: len(num),
'TotalPrice': lambda price: price.sum()})
```

```python
rfm
```

```python
rfm.columns
```

```python
#changing the columns names
#  
rfm.columns=['recency', 'frequency', 'monetary']
rfm
```

```python
rfm.head()
```

```python
#customers with lowest recency value and highest frequency & monetary to be considered as Best customer.
#Quantile statistical techniques is used to assign the score.
#in Quantiles techniques we make 4 equal parts based on available values.

```

```python
#computing the Quantile of RFM value for segmentation
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1', '2', '3', '4'])
rfm['f_quartile'] = pd.qcut(rfm['frequency'], 4, ['4', '3', '2', '1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4', '3', '2', '1'])
```

```python
rfm.head()
```

```python
#Now we can combine all three Quantile score in a single column, 
# and it will help us to create the logic base Customer segmentation as
#  per the business criteria.
#111 will be considered as the best performing customer as they have lowest recency and 
# highest frequency and monetary. 
#444 will be the worst segment

```

```python
    #### Combine RFM Quantile score
    rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
```
